```python
# Liste des libellés
libelles = ["Nom", "Âge", "Ville", "Profession"]

# Liste des valeurs correspondantes
valeurs = ["John Doe", 30, "Paris", "Ingénieur"]

# Affichage de la table
for libelle, valeur in zip(libelles, valeurs):
    print(f"{libelle}: {valeur}")
```

```python
from IPython.display import display
from tabulate import tabulate

# Données de la table
donnees = [
    ["Nom", "Âge", "Ville", "Profession"],
    ["John Doe", 30, "Paris", "Ingénieur"],
    ["Jane Smith", 25, "Londres", "Développeuse"],
    ["Alice Johnson", 35, "New York", "Avocate"]
]

# Générer la table HTML
table_html = tabulate(donnees, headers='firstrow', tablefmt='html')

# Afficher la table sans les balises HTML
display(table_html)


```
