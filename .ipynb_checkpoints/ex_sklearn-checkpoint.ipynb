```python
# Load the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectPercentile
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.tree import DecisionTreeRegressor

import seaborn as sns
```

```python
# Load the Phishing Email dataset
df = pd.read_csv('Phishing_Email.csv')
df.drop("Unnamed: 0", axis=1, inplace=True)
df.drop_duplicates(inplace=True)
```

```python
# Extract features and target variable
x = df['Email Text']
y = df['Email Type']
```

```python
target = df['Email Type'].replace({'Safe Email':1, 'Phishing Email':0})
sns.countplot(x=target, hue=target)
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Handle missing values by replacing NaN with an empty string
X_train = X_train.fillna('')
X_test = X_test.fillna('')
```

```python
# Vectorize the text data using TF-IDF
vectorizer = TfidfVectorizer(sublinear_tf=True, encoding='utf-8', decode_error='ignore', stop_words='english')
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Feature selection using SelectPercentile
selector = SelectPercentile(percentile=10)
X_train_selected = selector.fit_transform(X_train_vectorized, y_train)
X_test_selected = selector.transform(X_test_vectorized)

# Initialize Naive Bayes classifiers
gnb = GaussianNB()
bnb = BernoulliNB()
mnb = MultinomialNB()
```

```python
# Train the Naive Bayes classifiers
gnb.fit(X_train_selected.toarray(), y_train)
bnb.fit(X_train_selected, y_train)
mnb.fit(X_train_selected, y_train)

# Make predictions using the classifiers
y_pred_gnb = gnb.predict(X_test_selected.toarray())
y_pred_bnb = bnb.predict(X_test_selected)
y_pred_mnb = mnb.predict(X_test_selected)
```

```python
# Evaluate the classifiers
accuracy_gnb = accuracy_score(y_test, y_pred_gnb)
accuracy_bnb = accuracy_score(y_test, y_pred_bnb)
accuracy_mnb = accuracy_score(y_test, y_pred_mnb)

confusion_gnb = confusion_matrix(y_test, y_pred_gnb)
confusion_bnb = confusion_matrix(y_test, y_pred_bnb)
confusion_mnb = confusion_matrix(y_test, y_pred_mnb)
```

```python
print("Gaussian Naive Bayes Accuracy:", accuracy_gnb)
print("Bernoulli Naive Bayes Accuracy:", accuracy_bnb)
print("Multinomial Naive Bayes Accuracy:", accuracy_mnb)

print("Confusion Matrix (Gaussian Naive Bayes):\n", confusion_gnb)
print("Confusion Matrix (Bernoulli Naive Bayes):\n", confusion_bnb)
print("Confusion Matrix (Multinomial Naive Bayes):\n", confusion_mnb)

print("------------------------------------Explanation about the result------------------------------------")
```

The results indicate the performance of different Naive Bayes classifiers on the phishing email classification task. Here's how you can interpret the results:

1. **Gaussian Naive Bayes:**
   - Accuracy: 91.05%
   - Confusion Matrix:
     - True Positive (TP): 2201
     - False Positive (FP): 278
     - True Negative (TN): 993
     - False Negative (FN): 36
   - The Gaussian Naive Bayes classifier achieved an accuracy of 91.05%. It correctly classified 2201 emails as phishing emails (True Positives) and 993 emails as safe emails (True Negatives). It misclassified 278 safe emails as phishing emails (False Positives) and 36 phishing emails as safe emails (False Negatives).

2. **Bernoulli Naive Bayes:**
   - Accuracy: 97.06%
   - Confusion Matrix:
     - True Positive (TP): 2153
     - False Positive (FP): 19
     - True Negative (TN): 1252
     - False Negative (FN): 84
   - The Bernoulli Naive Bayes classifier achieved an accuracy of 97.06%. It correctly classified 2153 emails as phishing emails (True Positives) and 1252 emails as safe emails (True Negatives). It misclassified 19 safe emails as phishing emails (False Positives) and 84 phishing emails as safe emails (False Negatives).

3. **Multinomial Naive Bayes:**
   - Accuracy: 96.29%
   - Confusion Matrix:
     - True Positive (TP): 2228
     - False Positive (FP): 121
     - True Negative (TN): 1150
     - False Negative (FN): 9
   - The Multinomial Naive Bayes classifier achieved an accuracy of 96.29%. It correctly classified 2228 emails as phishing emails (True Positives) and 1150 emails as safe emails (True Negatives). It misclassified 121 safe emails as phishing emails (False Positives) and 9 phishing emails as safe emails (False Negatives).

In summary, all three Naive Bayes classifiers achieved relatively high accuracies in classifying phishing emails. The Bernoulli Naive Bayes classifier had the highest accuracy among the three. However, it's important to consider other metrics such as precision, recall, and F1-score to have a more comprehensive understanding of the classifier's performance.
